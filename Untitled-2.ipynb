```python
import yfinance as yf
import matplotlib.pyplot as plt
start = "2014-01-01"
end = '2022-10-01'

tcs = yf.download('TCS',start,end,progress=False)#TCS
lnvgy = yf.download('lnvgy',start,end,progress=False)#Infosys
wipro = yf.download('WIPRO.NS',start,end,progress=False)#WIPRO
blk = yf.download('BLK',start,end,progress=False)#BlackRock
hpq = yf.download('hpq',start, end, progress=False)


hpq["mean"]=hpq["Open"].mean()
print("hpq:")
print(hpq.head())
hpq["mean"].plot(label = "HP Mean")
hpq["Open"].plot(label = "HP Open")


lnvgy["mean"]=lnvgy["Open"].mean()
print("lnvgy:")
print(lnvgy.head())
lnvgy["mean"].plot(label = "Lenovo Mean")
lnvgy["Open"].plot(label = "Lenovo Open")
plt.legend()
plt.show()

```
