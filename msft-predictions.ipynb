```python

import yfinance as yf 
import matplotlib.pyplot as plt

start = "2014-01-01"
end = '2019-1-01'
msft = yf.download('MSFT',start,end,progress=False)
msft["RM50"]=msft["Open"].rolling(50).mean()
msft["RM200"]=msft["Open"].rolling(200).mean()
msft["RM50"].plot(label ="Rolline mean for Last 50 days")
msft["RM200"].plot(label ="Rolline mean for Last 200 days")
msft["Open"].plot(label= "Open prices")
plt.legend()
plt.title("Rolling mean predictions")
plt.show()
```
